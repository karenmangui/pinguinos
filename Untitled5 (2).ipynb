```python
#creamos una copia para empezar a limpiar los datos
df = penguins.copy()
```

```python
df['sex'] = df['sex'].replace({'Male': 0, 'Female': 1})

print(df)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
#Codificacimos los valores de especie isla y sexo en variables dumi

species_encoder = OrdinalEncoder()
island_encoder = OrdinalEncoder()
sex_encoder = OrdinalEncoder()
df[['species']] = species_encoder.fit_transform(df[['species']])
df[['island']] = island_encoder.fit_transform(df[['island']])
df[['sex']] = sex_encoder.fit_transform(df[['sex']])
```

```python
df= pd.DataFrame(df)
```

```python
from pandas.core.groupby.groupby import DataFrame
# localizamos los valores nulos
null_columns= df.columns[df.isnull().any()]
df[null_columns].isnull().sum()
```
